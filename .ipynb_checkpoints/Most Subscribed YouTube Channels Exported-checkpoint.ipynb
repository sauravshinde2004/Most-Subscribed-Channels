```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# reading dataset in program
data = pd.read_csv('Most Subscribed YouTube Channels_exported.csv')
```

```python
# printing first 5 line of data
data.head(5)
```

```python
# printing columns of dataset
data.columns
```

```python
# printing informaion of dataset
data.info()
```

```python
data.describe()
```

```python
# printing duplicate values avalable in dataset
data.duplicated().sum()
```

```python
# if there are duplicate value we can remove then using drop_duplicate() function
data.drop_duplicates()
```

```python
# printing shape of data 
data.shape
```

```python
# cleaning data
# finding null values form dataset 
data.isnull().sum()
```

```python
# if there any null value we can fill using fillna() function
data.fillna("NaN")
```

```python
# also we can remove null value using dromna() function
data.dropna()
```

```python
# Summary statistics
summary_stats = data['Subscribers (millions)'].describe()

# Histogram of subscribers
plt.figure(figsize=(12, 6))
sns.histplot(data['Subscribers (millions)'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Subscribers')
plt.xlabel('Subscribers (millions)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

summary_stats
```

```python
# Analyze the distribution of subscribers across different categories
plt.figure(figsize=(14, 7), facecolor='white')
category_counts = data['Category'].value_counts()
plt.bar(category_counts.index, category_counts.values, color='teal')
plt.title('Distribution of Subscribers Across Categories')
plt.xlabel('Category')
plt.ylabel('Number of Channels')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.show()

# Explore the relationship between subscribers and primary language
plt.figure(figsize=(14, 7), facecolor='white')
sns.boxplot(x='Primary language', y='Subscribers (millions)', data=data)
plt.title('Subscribers vs. Primary Language')
plt.xlabel('Primary Language')
plt.ylabel('Subscribers (millions)')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.show()

# Explore the relationship between subscribers and country
plt.figure(figsize=(14, 7), facecolor='white')
sns.boxplot(x='Country', y='Subscribers (millions)', data=data)
plt.title('Subscribers vs. Country')
plt.xlabel('Country')
plt.ylabel('Subscribers (millions)')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.show()
```

```python
# Aggregate data by category and calculate mean and median subscribers
category_agg = data.groupby('Category')['Subscribers (millions)'].agg(['mean', 'median'])

# Aggregate data by country and calculate mean and median subscribers
country_agg = data.groupby('Country')['Subscribers (millions)'].agg(['mean', 'median'])

# Visualize aggregated data for categories
plt.figure(figsize=(14, 7), facecolor='white')
category_agg.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Mean and Median Subscribers by Category')
plt.xlabel('Category')
plt.ylabel('Subscribers (millions)')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Metrics', labels=['Mean', 'Median'])
plt.grid(True)
plt.show()

# Visualize aggregated data for countries
plt.figure(figsize=(14, 7), facecolor='white')
country_agg.plot(kind='bar', color=['lightgreen', 'orange'])
plt.title('Mean and Median Subscribers by Country')
plt.xlabel('Country')
plt.ylabel('Subscribers (millions)')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Metrics', labels=['Mean', 'Median'])
plt.grid(True)
plt.show()
```

```python
# Correcting the column name and rerunning the ranking and visualization

data['Rank'] = data['Subscribers (millions)'].rank(method='max', ascending=False)

# Sort the dataframe by the rank
df_sorted = data.sort_values(by='Rank')

# Visualize the top 10 channels
plt.figure(figsize=(12, 8), facecolor='white')
df_sorted.head(10).plot(x='Name', y='Subscribers (millions)', kind='bar', legend=False, color='dodgerblue')
plt.title('Top 10 YouTube Channels by Subscribers')
plt.xlabel('Channel')
plt.ylabel('Subscribers (millions)')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.show()

# Display the sorted dataframe head for review
print(df_sorted.head())
```

```python

# Setting up the plot background to be white for better visibility
plt.figure(facecolor='white')

# Correlation matrix for the numerical variable
corr_matrix = data[['Subscribers (millions)']].corr()
print('Correlation Matrix:\
', corr_matrix)

# Scatter plot
sns.scatterplot(data=data, x='Subscribers (millions)', y='Subscribers (millions)')
plt.title('Scatter Plot of Subscribers (millions)')
plt.show()
```

```python
from sklearn.cluster import KMeans

# Selecting relevant features for clustering
df_cluster = data[['Subscribers (millions)', 'Brand channel', 'Category']]

# Encoding categorical data
df_cluster = pd.get_dummies(df_cluster)

# Applying KMeans Clustering
kmeans = KMeans(n_clusters=3, random_state=0).fit(df_cluster)
data['Cluster'] = kmeans.labels_

# Plotting the clusters
plt.figure(figsize=(10, 6))
plt.scatter(data['Subscribers (millions)'], data.index, c=data['Cluster'], cmap='viridis')
plt.title('Clusters of YouTube Channels')
plt.xlabel('Subscribers (millions)')
plt.ylabel('Channel Index')
plt.colorbar(label='Cluster')
plt.show()
print('Clustering done. Channels have been grouped based on their features.')
```
